```python
import serial
import time
import sys
```

```python
# Configuration - Change 'COM3' to your specific port
# (e.g., '/dev/ttyUSB0' on Linux, '/dev/cu.usbmodem14101' on macOS)
PORT = 'COM10'       # mine is COM5
BAUD_RATE = 9600

def run_servo_control():
    arduino = None
    try:
        # Initialize Serial Connection
        arduino = serial.Serial(port=PORT, baudrate=BAUD_RATE, timeout=1)
        # Wait for Arduino to reset after connection
        time.sleep(2)
        print(f"Successfully connected to Arduino on {PORT}")
        print("Type 'exit' to quit.")

        while True:
            user_input = input("Enter degrees (0-180): ").strip()

            if user_input.lower() == 'exit':
                break

            # 1. Validation: Check if numeric
            try:
                degrees = int(user_input)
            except ValueError:
                print("Invalid input! Please enter a whole number.")
                continue

            # 2. Validation: Check range
            if degrees < 0 or degrees > 180:
                print("Input out of bounds! Range is 0 to 180.")
                continue

            # 3. Transmission
            try:
                # Send value as string with newline
                arduino.write(f"{degrees}\n".encode('utf-8'))

                # Try to read response from Arduino
                response = arduino.readline().decode('utf-8').strip()
                if response:
                    print(f"Arduino says: {response}")

            except serial.SerialException as e:
                print(f"Connection lost: {e}")
                break

    except serial.SerialException as e:
        print(f"Error: Could not open port {PORT}. Check connection or port name.")
        print(f"Details: {e}")

    finally:
        # Crucial for Jupyter: Always close the port
        if arduino and arduino.is_open:
            arduino.close()
            print("Serial port closed.")

if __name__ == "__main__":
    run_servo_control()
```
