```python
import control as ctrl
import matplotlib.pyplot as plt
import numpy as np

# 1. Define the Transfer Function G(s)
num = [1, 1]
den = [1, 4, 7, 6, 3]
G = ctrl.TransferFunction(num, den)

print(f"Transfer Function: G(s) = {G}")

# 2. Determine Poles and Zeros
poles = ctrl.poles(G)
zeros = ctrl.zeros(G)

# 3. Generate Step Response
t, y = ctrl.step_response(G)

plt.figure(figsize=(10, 5))
plt.plot(t, y, color='#06b6d4', linewidth=2)
plt.title('Step Response')
plt.xlabel('Time (s)')
plt.ylabel('Output (x2)')
plt.grid(True, alpha=0.3)
plt.show()

# 4. Generate Pole-Zero Map
plt.figure(figsize=(6, 6))
ctrl.pzmap(G, grid=True)
plt.title('Pole-Zero Map')
plt.show()
```
