```python
import control as ctrl
import matplotlib.pyplot as plt
import numpy as np

# 1. Define the Transfer Function G(s) = 1 / (s^4 + 2s^3 + 3s^2 + 2s)
num = [1]
den = [1, 2, 3, 2, 0]
G = ctrl.TransferFunction(num, den)

print(f"Transfer Function: G(s) = {G}")

# 2. Determine Poles and Zeros
poles = ctrl.poles(G)
zeros = ctrl.zeros(G)
print(f"Poles: {poles}")
print(f"Zeros: {zeros}")

# 3. Generate Step Response Plot
t, y = ctrl.step_response(G)
plt.figure(figsize=(10, 6))
plt.plot(t, y, color='#0284c7', linewidth=2)
plt.title('Step Response Analysis')
plt.xlabel('Time (s)')
plt.ylabel('Output Amplitude')
plt.grid(True, alpha=0.3)
plt.show()

# 4. Generate Pole-Zero Map
plt.figure()
ctrl.pzmap(G, grid=True)
plt.title('System Pole-Zero Configuration')
plt.show()
```
