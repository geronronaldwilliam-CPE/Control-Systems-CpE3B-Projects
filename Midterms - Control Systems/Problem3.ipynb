```python
import control as ctrl
import matplotlib.pyplot as plt
import numpy as np

# 1. Define the Transfer Function G(s)
# Numerator: 4s^2 + 2s + 6
# Denominator: 16s^4 + 24s^3 + 52s^2 + 36s
num = [4, 2, 6]
den = [16, 24, 52, 36, 0]

G = ctrl.TransferFunction(num, den)
print(f"Transfer Function:\n{G}")

# 2. Determine Poles and Zeros
poles = ctrl.poles(G)
zeros = ctrl.zeros(G)
print(f"Poles: {poles}")
print(f"Zeros: {zeros}")

# 3. Generate Step Response Plot
t, y = ctrl.step_response(G)
plt.figure(figsize=(10, 5))
plt.plot(t, y)
plt.title('Step Response of the Mechanical System')
plt.xlabel('Time (s)')
plt.ylabel('Output')
plt.grid()
plt.show()

# 4. Visualize Poles and Zeros
plt.figure(figsize=(6, 6))
ctrl.pzmap(G, grid=True)
plt.title('Pole-Zero Map of G(s)')
plt.show()
```
