```python
from docx import Document
import os
from datetime import datetime

def create_word_doc(content, filename="generated.docx", folder="generated_docs"):
    os.makedirs(folder, exist_ok=True)

    filepath = os.path.join(folder, filename)

    doc = Document()

    # Title
    doc.add_heading("Generated Document", 0)

    # Content (split into paragraphs)
    for line in content.split("\n"):
        doc.add_paragraph(line)

    doc.save(filepath)

    return f"Document saved at: {filepath}"
```

```python
import json
import re
import os
from langchain_community.llms import OpenAI

print("Updating engine routing priority...")

llm = OpenAI(
    base_url="http://localhost:1234/v1",
    api_key="lm-studio",
    temperature=0.1
)

def word_tool(topic_or_content):
    try:
        import docx
    except ImportError:
        import subprocess, sys
        subprocess.check_call([sys.executable, "-m", "pip", "install", "python-docx"])
        import docx
        
    os.makedirs("generated_docs", exist_ok=True)
    doc = docx.Document()
    
    print("-> Asking local LLM to generate document text context...")
    content = llm.invoke(f"Write a comprehensive explanation or summary regarding: {topic_or_content}")
    
    doc.add_heading("Automated Report", level=1)
    doc.add_paragraph(content)
    
    filepath = os.path.join("generated_docs", "automated_output.docx")
    doc.save(filepath)
    return f"Success! Document created and saved at: '{filepath}'"

def run_agent_workflow(user_prompt):
    prompt_lower = user_prompt.lower()
    
    # FIX: Priority check. If the user wants to make/write a file, check that FIRST.
    if "write" in prompt_lower or "create" in prompt_lower or "generate" in prompt_lower or "docx" in prompt_lower:
        action = "Word Document Generator"
        action_input = user_prompt
    elif "search" in prompt_lower or "find" in prompt_lower or "turtle" in prompt_lower or "pdf" in prompt_lower:
        action = "Document Search"
        action_input = user_prompt
    else:
        action = "Final Answer"
        action_input = user_prompt

    print(f"\n[Agent Decision] Executing Tool: {action}")
    
    if action == "Document Search":
        return retrieve_docs(action_input)
    elif action == "Word Document Generator":
        return word_tool(action_input)
    else:
        return llm.invoke(user_prompt)

print("Priority Routing Engine is active!")
```

```python
import os
import re

def create_wiki_doc(prompt):
    """Fetches a clean summary from Wikipedia and writes a .docx file."""
    try:
        import wikipedia
        import docx
    except ImportError:
        return "Missing libraries! Please run: !pip install wikipedia python-docx"
    
    # ----------------------------------------------------
    # FIX 1: Set a custom User-Agent so Wikipedia doesn't block you
    # ----------------------------------------------------
    wikipedia.set_rate_limiting(True)
    # This tells Wikipedia you are a helpful research bot, preventing a blank block response
    wikipedia.wikipedia.headers = {
        'User-Agent': 'AI-Agent-Learning-Bot/1.0 (contact: your_email@example.com) BasePythonWikipedia/1.4.0'
    }

    # ----------------------------------------------------
    # FIX 2: Clean punctuation and aggressive filler words
    # ----------------------------------------------------
    # Convert to lowercase and strip trailing periods/question marks
    clean_prompt = prompt.lower().strip().rstrip('.? ')
    
    fillers = [
        "look up", "search for", "create a document of", 
        "what is", "what are", "about", "on wikipedia", 
        "and its applications", "its applications"
    ]
    
    for word in fillers:
        # Match words with optional spaces around them
        clean_prompt = re.sub(rf'\b{word}\b', '', clean_prompt)
    
    # Clean up any accidental double spaces left behind
    topic = " ".join(clean_prompt.split()).strip()
    
    # Fallback default if the string cleaning empties the prompt entirely
    if not topic:
        topic = "control system"
        
    print(f"📖 Searching Wikipedia for clean keyword: '{topic}'...")
    
    # Step C: Extract summary text safely from Wikipedia
    try:
        # Explicitly search for the page title first to get a highly relevant match
        search_results = wikipedia.search(topic)
        if not search_results:
            return f"❌ Wikipedia found absolutely no matches for '{topic}'."
            
        # Pick the top result from the search index
        best_match = search_results[0]
        summary = wikipedia.summary(best_match, sentences=4)
        
    except wikipedia.exceptions.DisambiguationError as e:
        # If the topic matches multiple items, cleanly pick the first alternative
        best_match = e.options[0]
        summary = wikipedia.summary(best_match, sentences=4)
    except Exception as e:
        return f"❌ Wikipedia API error for '{topic}'. Details: {str(e)}\nTip: Double check your internet connection or try again in a few seconds."
        
    # Step D: Create and write the document to disk
    try:
        os.makedirs("generated_docs", exist_ok=True)
        doc = docx.Document()
        
        doc.add_heading(f"Research Brief: {best_match.title()}", level=1)
        doc.add_paragraph("This summary report was compiled automatically by your AI Agent suite.")
        doc.add_paragraph(summary)
        
        # Build safe file layout names
        safe_filename = re.sub(r'[^a-zA-Z0-9]', '_', best_match)[:30]
        filepath = os.path.join("generated_docs", f"doc_{safe_filename}.docx")
        doc.save(filepath)
        
        return f"💾 Success! File written cleanly.\n📍 File Location: {filepath}\n\n--- Content Preview ---\n{summary}"
        
    except Exception as e:
        return f"❌ Failed to build or write the document file. Details: {str(e)}"
```
