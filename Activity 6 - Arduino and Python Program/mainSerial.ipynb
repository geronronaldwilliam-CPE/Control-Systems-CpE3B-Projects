```python
import serial        # library to communicate with Arduino via serial
import time          # library for delays

# Open serial connection to Arduino on COM7 at 9600 baud
arduino = serial.Serial(port='COM10', baudrate=9600, timeout=1)

# Wait 2 seconds to let Arduino reset after connection
time.sleep(2) 
```

```python
while True:
    # Ask user for input command
    command = input("Enter 1 to turn ON LED, 0 to turn OFF LED, q to quit: ")

    if command == '1':
        arduino.write(b'1')   # Send '1' to Arduino → turn LED ON
    elif command == '0':
        arduino.write(b'0')   # Send '0' to Arduino → turn LED OFF
    elif command == 'q':
        break                 # Exit loop if user enters 'q'
    else:
        print("Invalid command")  # Handle wrong input
```

```python
arduino.close()      # Close the serial connection when done
```
