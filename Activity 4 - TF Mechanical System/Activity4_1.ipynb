```python
import control as ctrl
import matplotlib.pyplot as plt

# Define transfer function G(s) = 1 / (s^2 + 2s + 2)
num = [1]          # numerator coefficients (constant 1)
den = [1, 2, 2]    # denominator coefficients (s^2 + 2s + 2)
G = ctrl.tf(num, den)   # create transfer function object
print(G)                # display transfer function
```

```python
# Compute and display poles of G(s)
poles = ctrl.poles(G)
print("Poles:", poles)
```

```python
# Compute and display zeros of G(s)
zero = ctrl.zeros(G)
print("Zero:", zero)
```

```python
# Plot pole-zero map with grid
import matplotlib.pyplot as plt

ctrl.pzmap(G, grid = True)
plt.show()
```
